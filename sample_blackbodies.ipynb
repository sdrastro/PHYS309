```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

plt.rcParams.update({'font.size':16, 'axes.labelsize':16})
```

```python
B_lam = lambda lam, T: (2*constants.pi * constants.c**2 * constants.h / lam**5) \
                       / (np.exp(constants.h * constants.c / (lam * constants.k * T)) - 1)
```

```python
micron2meter = 1e-6
```

```python
wavelengths = np.logspace(-1.5, 1, num=301, endpoint=True)
```

```python
plt.figure(figsize=(9,6))
plt.loglog(wavelengths, B_lam(micron2meter * wavelengths, 2000), color='sienna')
plt.loglog(wavelengths, B_lam(micron2meter * wavelengths, 3500), color='darkorange')
plt.loglog(wavelengths, B_lam(micron2meter * wavelengths, 5700), color='yellow')
plt.loglog(wavelengths, B_lam(micron2meter * wavelengths, 8000), color='mediumblue')
plt.xlabel('Wavelength (micron)')
plt.ylabel(r"Flux density (W / m$^3$)")
plt.ylim(1e-2, 1e16)
plt.grid(color='0.85')
```
