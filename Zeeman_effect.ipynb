# Zeeman effect

A state is <bf>degenerate</bf> when there is more than one wave function for a given energy (see Section 6.5 of your textbook). In the hydrogen atom, the energy levels $E_n$ depend <bf>only</bf> on the principal quantum number $n$, in agreement with the Bohr model. Wavefunctions with the same $n$ but different $\ell$ and $m_{\ell}$ have the same total energy. (For many-electron atoms, the situation is more complicated, and electrons with the same principal quantum number $n$ can have different energies.)

When an atom is placed in an external magnetic field, the degeneracy between the various $\ell$ and $m_{\ell}$ is broken. Instead of just one energy $E_n$ for the electron(s) with principal quantum number $n$, there are $2 \ell + 1$ available values of $E$, one for each value of $m_{\ell}$. The corresponding spectral line widens and splits.
