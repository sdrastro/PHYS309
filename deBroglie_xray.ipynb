Enter Pierre Louis de Broglie with the question, <em>can particles have wavelike behavior?</em>

From relativity we have <br/>
$E = pc$ (1)<br/>
for photons. We also know <br/>
$E = h \nu$. (2)<br/>
Set $pc = h \nu$. Next, recall that $c = \lambda \nu$. Sub into (1):<br/>
$E = p \lambda \nu$.<br/>
