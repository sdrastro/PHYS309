# Solutions to the time-independent Schrodinger wave equation: Infinite square well potential

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size':16, 'axes.labelsize':16})
```

```python
plt.figure(figsize=(5,5))
plt.arrow(0, 0, 0, 3, color='k', width=0.1)
plt.arrow(3, 0, 0, 3, color='k', width=0.1)
plt.fill_between([-1, 0], 0, 3.4, color='lightgray')
plt.fill_between([3, 4], 0, 3.4, color='lightgray')
plt.axhline(0, color='k')
plt.text(-0.2, 3.5, r"$\infty$", fontsize='x-large')
plt.text(2.8, 3.5, r"$\infty$", fontsize='x-large')
plt.text(-1.3, -0.1, '0', fontsize='large')
plt.text(-0.1, -0.27, '0', fontsize='large')
plt.text(2.9, -0.27, 'L', fontsize='large')
plt.xlim(-1, 4)
plt.ylim(-0.2, 3.8)
plt.xticks([])
plt.yticks([])
plt.box(False)
plt.xlabel('x', fontsize='x-large')
plt.ylabel('V(x)', fontsize='x-large')
```

Start with time-independent Schrodinger wave equation:

### $-\frac{\hbar^2}{2m} \frac{d^2 \psi(x)}{dx^2} + V(x) \psi(x) = E \psi(x)$

For $x \leq 0$ and $x \geq L$, $V(x) \rightarrow \infty$, so the only way to have a finite particle energy $E$ is if $\psi(x) = 0$. That means the particle <em>can't</em> get out of the well: it's constrained to be between $0$ and $L$.

Furthermore, the potential is defined so that $V(x) = 0$ for $0 < x < L$, so inside the box, the term $V(x) \psi(x)$ vanishes and the Schrodinger wave equation becomes

### $-\frac{\hbar^2}{2m} \frac{d^2 \psi(x)}{dx^2} = E \psi(x)$.         (1)

Dividing both sides of Equation (1) by $\hbar^2 / (2m)$ gives

### $\frac{d^2 \psi(x)}{dx^2} = \frac{-2mE}{\hbar^2} \psi(x)$.

Now we have the standard differential equation that describes the simple harmonic oscillator: $d^2 \psi(x) / dx^2 = -C^2 \psi(x)$, where $C$ is some constant. The solution is the free-particle wave equation, $\psi(x) = A e^{i k x}$. Let's double-differentiate:

### $\frac{d \psi(x)}{dx} = i k A e^{i k x}$

### $\frac{d^2 \psi(x)}{dx^2} = i^2 k^2 A e^{i k x} = -k^2 A e^{i k x}$.

How can we find the constants, $k^2$ and $A$? First, consider the particle's energy. Inside the box with $V = 0$, all is kinetic and none is potential, so $E = K = \frac{1}{2} m v^2$. Then we rearrange the definition of momentum, $p = mv$, to get $v = p/m$ and substitute for $v$ in the equation for kinetic energy, which gives

### $E = \frac{1}{2} m \left( \frac{p}{m} \right)^2 = \frac{p^2}{2m}$.     (2)

Next we go back to de Broglie: $p = \frac{h}{\lambda} = \hbar k$. Substituting in for $p$ in Equation 2 gives

### $E = \frac{\hbar^2 k^2}{2m}$.

Now we can substitute in for $E$ on the right-hand side of Equation (1):

### $-\frac{\hbar^2}{2m} \frac{d^2 \psi(x)}{dx^2} = \frac{\hbar^2 k^2}{2m} \psi(x)$,

which reduces to

### $\frac{d^2 \psi(x)}{dx^2} = -k^2 \psi(x)$.

Now we use Euler's formula,

### $e^{i k x} = \cos(kx) + i \sin(kx)$.

Recall that our solution to the Schrodinger wave equation must be continuous for all $x$, including at the boundaries of the box. (The requirement that its first derivative $d\psi(x)/dx$ must also be continuous everywhere is waived at the boundaries since the potential goes up to infinity.) But $\cos(kx)$ is 1 at $x = 0$, so we have to get rid of the cosine part of our wave function: even though it satisfies the differential equation (1), it doesn't satisfy the boundary conditions. Our solution is now

### $\psi(x) = A i \sin (kx)$     (3).

Let's set $\psi(x) = 0$ at $x = L$. Any value of $k$ which makes the sine function have a zero crossing at $x = L$ satisfies the right-hand boundary condition. The sine function has zero crossings at integer multiples of $\pi$, so $\psi(x) = 0$ wherever $k x = n \pi$. At $x = L$, we have $k L = n \pi$, which gives

### $k = \frac{n \pi}{L}$.

Our solution to the Schrodinger equation for the infinite square well potential is

### $\psi(x) = A \sin \left( \frac{n \pi x}{L} \right)$,

where we have dropped the imaginary number from Equation (3) because the probability function from Equation (3), $\psi^*(x) \psi(x)$, is $A^2 [-i \sin (kx)] [(i \sin (kx)] = -A^2 i^2 \sin^2 (kx) = A^2 \sin^2 (kx)$. We get the exact same probability function from $\psi(x) = A \sin (kx) = A \sin (n \pi x / L)$.

Now for the value of $A$: we get it from the probability function. We know

### $\int_{-\infty}^{\infty} \psi^*(x) \psi(x) dx = 1$,

and since $\psi(x) = 0$ for $x \leq 0$ and $x \geq L$, we can change the bounds of our integral to get

### $\int_{0}^{L} \psi^*(x) \psi(x) dx = \int_{0}^{L} A^2 \sin^2 \left( \frac{n \pi x}{L} \right)dx = 1$ .

The integral of $\sin^2 (\theta)$ is $\frac{1}{2} \theta - \frac{1}{4} \sin (2 \theta)$. Set $\theta = \frac{n \pi x}{L}$, $x = \frac{L \theta}{n \pi}$ and $dx = \frac{L d \theta}{n \pi}$ so we can rewrite our integral:

### $A^2 \int_{\theta = 0}^{n \pi} \sin^2 (\theta) \frac{L}{n \pi} d\theta = \left. \frac{A^2 L}{n \pi} \left[ \frac{1}{2} \theta - \frac{1}{4} \sin (2 \theta) \right] \; \right|_{\theta = 0}^{n \pi} = 1$.

For $\theta = 0$ the whole expression in brackets reduces to zero, so just worry about the upper limit of integration, which gives

### $\frac{A^2 L}{n \pi} \frac{n \pi}{2} - \frac{1}{4} \sin (2 n \pi) = 1$.

The second term above reduces to zero, so the whole expression becomes $\frac{A^2 L}{2} = 1$ and

### $A = \sqrt{\frac{2}{L}}$.

The complete wave function is

## $\psi(x) = \sqrt{\frac{2}{L}} \sin \left( \frac{n \pi x}{L} \right)$ for integer $n \geq 1$.

The lowest-energy state or "ground state" of the infinite square well potential has $n = 1$</em>. Higher $n$ corresponds to higher energy.

```python
x = np.linspace(0, 1, num=101, endpoint=True)
psi = lambda x, n: np.sin(n*np.pi*x) # L = 1 

fig, axarr = plt.subplots(4, 1, figsize=(6.5,7.5), sharex=True)

for i, ax in enumerate(axarr):
    ax.plot(x, psi(x, i+1), color='mediumblue', label='n = '+f"{i+1}")
    ax.legend(loc='best')
    ax.axhline(0, color='k')
    ax.set_yticks([])
    ax.fill_between([-0.1, 0], [-1, -1], [1, 1], color='lightgray')
    ax.fill_between([1, 1.1], [-1, -1], [1, 1], color='lightgray')
    
axarr[0].set_title(r"$\psi(x)$", fontsize='x-large')
axarr[3].set_xticks([0, 1])
axarr[3].set_xticklabels(["0", "L"])
axarr[3].set_xticks([0, 1])
axarr[3].set_xlabel('x', fontsize='x-large')
```
