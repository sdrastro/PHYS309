Recall that $\vec{D} = \epsilon_0 \vec{E} + \vec{P}$ is the electric displacement field, where $\vec{P}$ is the polarization density in the material. For air, $\vec{P} \approx 0$ and $\vec{D} = \epsilon_0 \vec{E}$. 

### Experiment goal: suspend oil drop between plates

Doing so requires balancing electrical and gravitational forces: $q \vec{E} + m \vec{g} = 0$.

If $V$ is voltage across plates, then $E = \frac{V}{d}$ and $q = \frac{m g d}{V}$. But we still don't know $m$.

### Millikan found the mass of the oil drops by watching them reach terminal velocity after he turned the electric field off. 

With no more upward-directed electrical force, the drops would fall at an accelerating rate until they reached terminal velocity, at which $m\vec{g} + \vec{F}_d = 0$ (where $\vec{F}_d$ is the drag force).

The drag force is given by $\vec{F}_d = -\frac{1}{2} C \rho A u^2 \hat{u}$, where $C$ is the drag coefficient, $\rho$ is the density of air, $A = \pi r^2$ is the cross-sectional area of the roughly spherical air drop, $u$ is terminal speed, and $\hat{u}$ is a unit vector that points in the direction of the oil drop's motion (straight downward in this case).

If the air is "quiet" (low turbulence) and the oil drop moves slowly, $C \propto 1/u$ and the drag law simplifies to $\vec{F}_d = 6 \pi \mu r u$, where $\mu$ is the dynamic viscosity of air.

$\mu$ was known and $r$ and $u$ could be measured in Millikan's setup. With known $F_d$, $m$  and finally $q$ were measurable.
