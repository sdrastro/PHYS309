# Introducing the Schrodinger wave equation

This equation is foundational to quantum mechanics. For some simple <em>potentials</em> (to be defined below), there is an analytic solution. For more complicated potentials like a many-electron atom or molecule, there is not. Atomic and molecular physicists build simplifying approximations for complicated potentials write codes for numerically solving the Schrodinger wave equation.

### First, an introduction to complex exponentials

Recall that our treatment of traveling waves has been based on sinusoids of the form

### $\Psi(x,t) = A \sin(kx - \omega t)$. $\; \; \; \;$ (1)

We can build valid probability functions $\Psi^* (x,t) \Psi(x,t)$ by adding together many sinusoids:

### $\Psi(x,t) = \sum_{i=1}^N A_i \sin(k_i x - \omega_i t)$. $\; \; \; \;$ (2)

If our particle can be anywhere on the x-axis, we require $N = \infty$. If the particle is <em>confined</em> by a potential (more on this later), $N$ can be finite, or even 1.

Equations (1) and (2) are not quite complete, however. They don't offer us a choice of <em>phase</em>. To make Equation (1) fully general, we need to add a phase term:

### $\Psi(x,t) = A \sin(kx - \omega t + \phi)$.  $\; \; \; \;$ (3)

Likewise, we adjust Equation (2) by adding a phase term $\phi_i$ to the argument of each sinusoid.

However, there is an equivalent way to express Equation (3):

### $\Psi(x,t) = A \cos(kx - \omega t) + B \sin(kx - \omega t)$.  $\; \; \; \;$ (4)

We can find a combination of $A$ and $B$ that will replicate Equation (3) for any value of $\phi$.

#### The cleverness of complex exponentials

Now suppose we have a function of the form

### $\Psi(x,t) = (A + Bi) e^{i(kx - \omega t)}$, $\; \; \; \;$ (5)

where $i$ is the imaginary number. Euler's identity says $e^{i(kx - \omega t)} = \cos(kx - \omega t) + i \sin(kx - \omega t)$. (You can verify this yourself by writing out the power series expansions of $\cos(\theta)$, $\sin(\theta)$, and $e^{\theta}$.) Equation (5) is a compact version of Equation (4). Once again, there is a combination of amplitudes $A$ and $B$ that can represent a wave with any phase. (Note that your textbook expresses $A + Bi$ as a single complex constant $A$.)

I hope Equation (5) will convince you that <strong>complex exponentials aren't scary.</strong> They're just a clever way to express the fact that our wave functions may have both cosine and sine components. Cosines are encoded in the real part of the complex exponential and sines are encoded in the imaginary part.

#### Isn't it weird that our wave function can be complex? Who ever heard of a complex value of momentum or energy?

A complex-valued $\Psi(x,t)$ is not a problem! That's because the wave function <em>itself</em> is not measurable. Only the physically measurable quantities such as position, momentum, and energy are required to be real-valued, and those are determined by the probability function. Defining the probability function as $P(x,t) = \Psi(x,t)^* \Psi(x,t)$ ensures that our observables will be real-valued!!

### The Schrodinger wave equation

Recall that we derived a simple partial differential equation for a wave traveling in one dimension:

$\frac{\partial^2 \Psi(x,t)}{\partial t} = -v^2 \frac{\partial^2 \Psi(x,t)}{\partial x}$

where $v$ is the phase velocity. The Schrodinger wave equation is slightly different: the time derivative is first-order and gets multiplied by the imaginary number.

## $i \hbar \frac{\partial \Psi(x,t)}{\partial t} = -\frac{\hbar^2}{2m} \frac{\partial^2 \Psi(x,t)}{\partial x^2} + V(x,t) \Psi(x,t)$. $\; \; \; \;$ (6, Schrodinger) 

When you look at Equation (6), don't think "$i$ is scary." Instead, think "$i$ represents sines." 

In Equation (6), $m$ is the particle mass and $V(x,t)$ is the time- and space-dependent <em>potential</em> that confines the particle. You already understand potentials. One example is Earth's gravitational potential, which keeps us, the atmosphere, orbiting satellites, and the moon confined. If we want to escape Earth's gravitational potential, we need to blast off the surface at a speed exceeding $v_{\rm esc} = \sqrt{2 G M_{\oplus} / R_{\oplus}}$. (In astronomy, the symbol $\oplus$ refers to the earth.)

Confusingly, although we call $V(x,t)$ a potential, it actually has energy units, and should be understood as the minimum energy the particle must have in order not to be confined (though tunneling is possible for a confined particle, as we will see).

### Valid wave functions

A valid wave function must of course satisfy Equation (6), but there are other requirements:

<ul>
    <li>$\Psi(x,t)$ must be finite for all $(x,t)$. Otherwise the probability function would integrate up to infinity, which is nonsense. (Note that $\Psi(x,t) = \delta(x_0)$ is invalid because of the Heisenberg uncertainty principle, which we will learn about next time.)</li>
    <li>$\Psi(x,t)$ must be single-valued for all $(x,t)$. If the $\Psi(x,t)$ is not single-valued, there may be multiple values of the probability for some $(x,t)$. So none of this:</li>
</ul>

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size':16, 'axes.labelsize':16})
```

```python
Psi = lambda x: np.sqrt(x)
xx = np.linspace(0, 30, num=301)
plt.plot(xx, Psi(xx), color='k')
plt.plot(xx, -Psi(0.7*xx), color='k')
plt.axhline(0, color='k', ls=':')
plt.axvline(0, color='k', ls=':')
plt.xlabel('x (arbitrary units)')
plt.ylabel(r"$\Psi(x, t=0)$")
plt.grid(color='0.85')
plt.xlim([-30, 30])
plt.ylim([-6, 6])
```

<ul>
    <li><strong>If $V(x,t)$ is finite</strong>, $\Psi(x,t)$ and $\partial \Psi(x,t) / \partial x$ must both be continuous. Otherwise $\partial^2 \Psi(x,t) / \partial x^2$ could be undefined at some $(x,t)$, in turn making $E$ and $p$ undefined. So none of this either:</li>
</ul>

```python
plt.plot(xx, xx, color='k')
plt.plot(-xx, xx, color='k')
plt.scatter([0], [0], color='r', marker='o', s=100)
plt.xlabel('x (arbitrary units)')
plt.ylabel(r"$\Psi(x, t=0)$")
plt.text(-14, -3, r"Undefined $p$ and $E$", color='r')
plt.grid(color='0.85')
plt.ylim([-4, 31])
```

(If $V(x,t)$ is infinite for some $x$, then $\Psi(x,t)$ and $\partial \Psi(x,t) / \partial x$ may both be discontinuous at that value of $x$. This situation is not realizable in physical reality, but it is a useful approximation in some cases.)

<ul>
    <li>In order to normalize the probability function, $\Psi(x,t)$ must approach zero as $x \rightarrow \infty$ and $x \rightarrow -\infty$. Otherwise the normalization constant $A$ is a nonsensical $1/\infty$.</li>
    <li>(Your book doesn't have this one, but I'm adding it.) It must be true that $\int_{-\infty}^{\infty} \Psi(x,t)^*\Psi(x,t) dx = 1$ for all times $t$. There are some functions, such as <br/>$\Psi(x,t=0) = 1/x$ for all $x > 1$, otherwise $0$,<br/> that approach zero as $x \rightarrow \infty$ but still have $\int_{-\infty}^{\infty} \Psi(x,t)^*\Psi(x,t) dx = \infty$. </li>
