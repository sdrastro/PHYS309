# Gaussian wave packet

## $G(x) = \frac{1}{\sigma \sqrt{2 \pi}} \exp \left[ -{\frac{(x - \mu)^2}{2 \sigma^2}} \right]$

$\sigma$ = standard deviation, $\mu$ = mean

### Q: As $\sigma \rightarrow 0$, does the particle get more or less localized?

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size':16, 'axes.labelsize':16})
```

```python
Gaussian = lambda x, mu, sigma: np.exp(-(x-mu)**2 / (2*sigma**2)) / \
                                (sigma * np.sqrt(2*np.pi))

x = np.linspace(0, 10, num=101, endpoint=True) - 5
plt.plot(x, Gaussian(x, 0, 1), color='mediumblue', lw=2)
plt.xlabel('x')
plt.ylabel(r"$G(x)$")
plt.grid(color='0.85')
```

### Now for the wave part

Recall the definition of wave number: 

### $k = \frac{2 \pi}{\lambda}$

Likewise, recall that arguments to exponentials have to be dimensionless. We can use a <em>wave number width $\Delta k$</em> in place of $\sigma$ to specify which wavelengths will be part of our wave function. The equation is

### $\psi(x) = A \exp \left[ -\Delta k^2 \Delta x^2 \right] \cos(k_0 x)$,

where $k_0$ is the wavenumber at the "center" of the wave packet.

```python
Gaussian_wave = lambda x, delta_k, k0: np.exp(-(x**2 * delta_k**2)) * np.cos(k0*x)

k0 = 2*np.pi
plt.plot(x, Gaussian_wave(x, 0.5, k0), color='mediumblue', lw=2)
plt.xlabel('x')
plt.ylabel(r"$\psi(x)$")
plt.grid(color='0.85')
```
