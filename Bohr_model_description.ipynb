# Bohr's quantum model of the hydrogen atom

Rutherford's model explains the results of the $\alpha$ particle scattering experiments, but it has a flaw: electrons would spiral into atomic nuclei within about $10^{-9}$ seconds. The classical atomic model based on planets orbiting the sun produces unstable atoms. We need a way to keep atoms stable for long periods of time.

### Bohr's proposal: It's possible for electrons to move around nuclei <em>without</em> emitting photons

This behavior would violate classical electromagnetic theory, but then again, we have already learned about several phenomena that violate classical theory: Compton scattering, the photoelectric effect, and the idea that light is "quantized" - you have to emit or absorb a whole photon, whereas in classical physics a wave can transfer part of its energy to something else.

Bohr based his new model of the atom on four assumptions that were consistent with experimental results. He then used those assumptions to derive a new, quantum model of the hydrogen atom. Bohr followed a similar process to Einstein's derivation of special relativity from just two postulates.

#### Bohr's assumptions:

<ol>
    <li>Atoms have "stationary states", which we now describe as energy levels. As long as an electron stays within its level, it doesn't radiate light.</li>
    <li>Light emission or absorption only occurs when an electron moves <em>between</em> energy levels. Since (a) energy levels are well defined, and (b) it's not possible to absorb or emit a fraction of a photon, atoms can only absorb/emit photons with energy $h\nu = E_{\rm high} - E_{\rm low}$, where $E_{\rm high}$ is a higher energy level and $E_{\rm low}$ is a lower level.$^{\star}$</li>
    <li>As long as the electrons are in their energy levels and not moving between them, classical physics concepts can explain the dynamics of the atom. For example, you can calculate a binding energy and a magnetic moment - concepts from PHYS 207 and 208. But the electron's motion between energy levels cannot be described by classical physics.</li>
    <li>The kinetic energy of the electron-nucleus system is quantized. $K$ can't take on just any value, but instead is restricted to $K = n h \nu_{\rm orb} / 2$, where $n$ is an integer, $h = 6.6261 \times 10^{-34}$ J/Hz is Planck's constant, and $\nu_{\rm orb}$ is the frequency of the electron's orbit around the nucleus in Hz. Likewise, the electron orbit's angular momentum cannot take on any value, but is restricted to $L = n \hbar$ (where $\hbar = \frac{h}{2 \pi}$).</li>
</ol>

Assumption 4 was the hardest for Bohr's contemporaries to accept. That's understandable, since in gravitation and classical electrodynamics the kinetic energy of an orbiting planet / moving charge is measured on a continuum. But since Einstein had already provided a quantum framework for the energy of light (photons), there was a precedent for a quantum treatment of kinetic energy. 

$^{\star}$In many-electron atoms, not all transitions are equally likely. Some are highly probable and some are so rare as to be called "forbidden." You will learn more about the likelihood of various transitions in PHYS 424.
