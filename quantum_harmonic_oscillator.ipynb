# Quantum harmonic oscillator

In PHYS 207, you learned about simple harmonic motion, which describes a block on a spring. Harmonic motion exists in the quantum world as well, for example in vibrational modes of molecules.

```python
from PIL import Image
vibr = Image.open("vibrational_modes.jpeg")
display(vibr)
```

From Hooke's law, $F = -\kappa (x - x_0)$ (where $\kappa$ is the spring constant and $x_0$ is the equilibrium position), the harmonic oscillator potential energy is $V = \frac{1}{2} \kappa (x - x_0)^2$.

Starting with the time-independent Schrodinger wave equation,

### $\frac{-\hbar^2}{2m} \frac{d^2 \psi(x)}{dx^2} + V(x) \psi(x) = E \psi(x)$,

we sub in our expression for $V(x)$ and assume $x_0 = 0$ to get

### $\frac{-\hbar^2}{2m} \frac{d^2 \psi(x)}{dx^2} + \frac{1}{2} \kappa x^2 \psi(x) = E \psi(x)$

### $\frac{-\hbar^2}{2m} \frac{d^2 \psi(x)}{dx^2} = E \psi(x) - \frac{1}{2} \kappa x^2 \psi(x)$

### $\frac{-\hbar^2}{2m} \frac{d^2 \psi(x)}{dx^2} = (E - \frac{1}{2} \kappa x^2) \psi(x)$

### $\frac{d^2 \psi(x)}{dx^2} = -\frac{2m}{\hbar^2} (E - \frac{1}{2} \kappa x^2) \psi(x)$

Now set $\alpha^2 = m \kappa / \hbar^2$ and $\beta = 2mE / \hbar^2$ and sub in to get

### $\frac{d^2 \psi(x)}{dx^2} = (\alpha^2 x^2 - \beta) \psi(x)$.

This differential equation is different than our previous experience with the Schrodinger wave equation, which always gave $\frac{d^2 \psi(x)}{dx^2} = C^2 \psi(x)$, where $C$ is a positive or negative constant. Here we have a term that's proportional to $x^2 \psi(x)$. The solution is

### $\psi(x) = H_n(x) e^{-\alpha x^2/2}$,

where $H_n(x)$ is a constant multiplied by the $n^{\rm th}$ [Hermite polynomial](https://mathworld.wolfram.com/HermitePolynomial.html). The $e^{-\alpha x^2/2}$ part of $\psi(x)$ is a Gaussian.

### Can the particle have zero kinetic energy?

Here is a graph of the potential:

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size':16, 'axes.labelsize':16, 'font.family':'sans-serif'})
```

```python
x = np.linspace(-1, 1, 101, endpoint=True)
plt.figure(figsize=(5,5))
plt.plot(x, x**2, color='k')
plt.xlabel('x (arbitrary units)')
plt.ylabel('V(x) (arbitrary units)')
plt.axvline(0, color='k', ls='--')
plt.grid(color='0.85')
```

From our graph, we see that the particle cannot have zero kinetic energy. Since $V(x) \rightarrow \infty$ for $x \rightarrow \pm \infty$, the particle is trapped in the well and cannot get out (it cannot "jump" over an infinite potential wall). That means the simple harmonic oscilllator is a bound system with $E \leq V$. Therefore, the only possible way to get $E = 0$ would be if $V = 0$, which happens only at $x = 0$. For the zero-energy particle, we would know the $x$ coordinate exactly. But if $E = 0$, then $E = p^2 / (2m)$ gives $p = \sqrt{2mE} = 0$, and now we know both the position and momentum exactly, which is forbidden by the uncertainty principle.

### Can we estimate the minimum particle energy?

Yes, we can do this without knowing $H_n(x)$. Recall from PHYS 207 that the <em>average</em> kinetic and potential energies of a simple harmonic oscillator are $\langle K \rangle = \langle V \rangle = \frac{1}{2} (K + V) = \frac{1}{2}$ (total energy). For a quantum particle with total energy $E$,

### $\langle K \rangle = \frac{1}{2}E = \frac{1}{2} \kappa \langle x^2 \rangle = \frac{\langle p^2 \rangle}{2m}$    (1)

### $\kappa \langle x^2 \rangle = \frac{\langle p^2 \rangle}{m}$

The particle's mean <em>deviation</em> from equilibrium, $\Delta x$, is the square root of the mean squared deviation $\langle x^2 \rangle$. Likewise, the deviation of the momentum from zero is $\Delta p = \sqrt{\langle p^2 \rangle}$, so

### $\Delta x = \frac{\Delta p}{\sqrt{\kappa m}}$    (2)

Given the Heisenberg uncertainty principle, $\Delta x \Delta p \geq \hbar/2$, we have 

### $\Delta x \geq \frac{\hbar}{2 \Delta p}$    (3).

Now we are going to rewrite $\langle x^2 \rangle$ as the product of Equations (2) and (3) and substitute into (1):

### $E \geq \kappa \frac{\Delta p}{\sqrt{\kappa m}} \frac{\hbar}{2 \Delta p}$

### $E \geq \frac{\hbar}{2} \sqrt{\frac{\kappa}{m}}$

Given that for a simple harmonic oscillator, $\omega = \sqrt{\kappa / m}$, the minimum energy of the simple harmonic oscillator is

## $E \geq \frac{\hbar \omega}{2}$.
